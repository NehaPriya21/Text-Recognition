# Creating Bounding box on image using search term

```python
def read_image(im):
    im3 = PIL.Image.open(im)
    bounds=reader.readtext(im)
    return im3, bounds
 
def plot_image(im):
  im3 = PIL.Image.open(im)
  imgplot = plt.imshow(im3)
  plt.show()

plot_image("opencv.jpg")
```

```python
def draw_boxes(search_term,image,bounds,color='Red',width=2):
  draw =ImageDraw.Draw(image)
  for bound in bounds:
      if re.match(search_term,bound[1]):
          p0, p1, p2, p3=bound[0]
          draw.line([*p0,*p1,*p2,*p3,*p0],fill=color,width=width)
  return image
```

```python
search_term = "Neha"
im3, bounds = read_image("opencv.jpg")
draw_boxes(search_term,im3,bounds)
```

```python
plot_image("vision3.png")
```

```python
search_term = "Non ADR"
im3, bounds = read_image("vision3.png")
draw_boxes(search_term,im3,bounds)
```

# Selecting portion of image to extract text

```python
import cv2
```

```python
def cordim(list):
    im = cv2.imread(list)
    roi = cv2.selectROI(im)
    im_cropped = im[int(roi[1]):int(roi[1]+roi[3]),
                int(roi[0]):int(roi[0]+roi[2])]
    cv2.imshow("Cropped Image", im_cropped)
    cv2.waitKey(0)
    cv2.destroyAllWindows() 
    image_roi= im[int(roi[0]):int(roi[3]),int(roi[1]):int(roi[2])]
    return image_roi
```

```python
import pytesseract
pytesseract.pytesseract.tesseract_cmd = 'C:\Program Files\Tesseract-OCR\\tesseract.exe'

bound = cordim('vision.jpg')
text = pytesseract.image_to_string(bound)
text
```
